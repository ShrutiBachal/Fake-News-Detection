```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
```

```python
true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
true['label'] = 1
fake['label'] = 0
```

```python
news = pd.concat([fake,true],axis=0)
news.head()
news.tail()
```

```python
news = news.drop(['title','subject','date'],axis = 1)
```

```python
news.head()
news.tail()
```

```python
news = news.sample(frac=1)    # Shuffling the dataset
news.head()
```

```python
news.reset_index(inplace=True)
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)  # axis = 1 means column
```

```python
news.head()
```

```python
import re     # to make it machine readable
```

```python
def wordpot(text):
  #Convert into lowercase
  text = text.lower()

  #Remove HTML tags
  text = re.sub(r'<.*?>','',text)

  #Remove URLs
  text = re.sub(r'https?://\S+|www\.\S+','',text)

  #Remove punctuation
  text = re.sub(r'[^\w\s]','',text)

  #Remove digits
  text = re.sub(r'\d','',text)

  #Remove newline characters
  text = re.sub(r'\n','',text)

  return text
```

```python
news['text'] = news['text'].apply(wordpot)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape   # this size matches to y_train because y is label to x i.e., y depends on x
```

```python
y_train.shape
```

```python
x_test.shape    # test and train divided on the basis of 0.3
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr = LR.predict(xv_test)  #check for test
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train,y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordpot)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_LR = LR.predict(new_xv_test)
  pred_rfc = rfc.predict(new_xv_test)
  pred_DTC = DTC.predict(new_xv_test)
  #pred_gbc = gbc.predict(new_xv_test)
  return "LR Prediction: {}  RFC Prediction: {}   DTC Prediction: {}".format(output_lable(pred_LR[0]),output_lable(pred_rfc[0]),output_lable(pred_DTC[0]))
```
